#### Some common Activation functions

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

x = np.linspace(-5, 5, 1000)
activation_functions = [
    ('Sigmoid', lambda x: 1 / (1 + np.exp(-x)), 'Pro: Smooth gradient', 'Con: Vanishing gradient'),
    ('Tanh', np.tanh, 'Pro: Centered at zero', 'Con: Vanishing gradient'),
    ('ReLU', lambda x: np.maximum(0, x), 'Pro: Fast computation', 'Con: Dead neurons'),
    ('Leaky ReLU', lambda x: np.maximum(0.1 * x, x), 'Pro: Solves dead neuron issue', 'Con: No standard value for leak'),
    ('Softmax', lambda x: np.exp(x) / np.sum(np.exp(x)), 'Pro: Probability distribution', 'Con: Computationally expensive'),
    ('ELU', lambda x: x * (x > 0) + (np.exp(x) - 1) * (x <= 0), 'Pro: Smooth gradient', 'Con: Computationally expensive')
]

fig, axs = plt.subplots(2, 3, figsize=(15, 10))
fig.tight_layout(pad=3.0)

for i, (name, func, pro, con) in enumerate(activation_functions):
    ax = axs[i // 3, i % 3]
    ax.plot(x, func(x), linewidth=2, color='black')
    ax.set_title(name)
    green_patch = mpatches.Patch(color='green', label=pro)
    red_patch = mpatches.Patch(color='yellow', label=con)
    ax.legend(handles=[green_patch, red_patch], loc='upper left')
    ax.set_facecolor('#F0F0F0')
    ax.set_title(name, fontweight='bold')
    for spine in ax.spines.values():
        spine.set_linewidth(2)


plt.show()


```

    The binary cross-entropy loss function, also known as log loss, is a fundamental concept in machine learning, specifically in the domain of classification problems. It is highly relevant in today's data-driven world, as it provides a quantifiable measure of the performance of binary classification models, such as logistic regression and neural networks, on real-world problems like spam detection, medical diagnosis, and fraud prevention. As a critical aspect of model evaluation and optimization, understanding binary cross-entropy loss is essential for professionals in fields such as data science, machine learning engineering, and artificial intelligence research.

    In practice, the binary cross-entropy loss function measures the dissimilarity between the true labels (ground truth) and the predicted probabilities generated by a classification model. It is defined as L(y, p) = -[y * log(p) + (1-y) * log(1-p)], where y represents the true label and p denotes the predicted probability. By minimizing this loss function, machine learning practitioners can enhance their models' ability to generalize and make accurate predictions on unseen data. The binary cross-entropy loss function is widely used in deep learning frameworks, making it a vital skill for job roles such as deep learning engineers and AI researchers.

```python
import numpy as np

x1 = np.array([0, 1, 2, 3])
y = np.array([0, 1, 0, 0])

def twoLayerNetwork(x):
    Wh = np.array([2, 2])
    Wo = np.array([-1, 1])

    bh = np.array([-2, -1])
    bo = np.array([-.5])

    n = len(x)

    # Hidden Scores
    zh = np.zeros((n, 2))
    for i in range(n):
        zh[i] = Wh * x[i] + bh

    # Hidden Activations
    ah = np.zeros((n, 2))
    for i in range(n):
        for j in range(2):
            ah[i, j] = 1 if zh[i, j] > 0 else 0

    # Output Scores
    zo = np.zeros(n)
    for i in range(n):
        zo[i] = Wo @ ah[i] + bo

    # Output Activations
    ao = np.zeros(n)
    for i in range(n):
        ao[i] = 1 if zo[i] >= 0 else 0

    print(ao)
    return ao

y = twoLayerNetwork(x1)
print('Part a of the extra credit is satisfied by the output above.')

# test the output from x = -5 to 5 with step size 0.05
x2 = np.arange(-5, 5, 0.05)
y = twoLayerNetwork(x2)
# plot x2 vs y
import matplotlib.pyplot as plt
plt.plot(x2, y)
plt.show()

#find the minimum x2 where y = 1
x2_min = x2[np.where(y == 1)]
print('The minimum x2 where y = 1 is: ', x2_min)


twoLayerNetwork(np.array([.51]))





```
