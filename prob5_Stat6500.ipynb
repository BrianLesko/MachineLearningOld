##### a) We are given n = 7 observations in p = 2 dimensions. For each observation, there is an associated class label, Sketch the Data

```python
# first we import the necessary libraries

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import sklearn
from sklearn import linear_model, preprocessing
```

```python
#Next, transcribe the data
# x1 is a 7 by 1 array
x1 = np.array([3, 2, 4, 1, 2, 4, 4])
# x2 is a 7 by 1 array
x2 = np.array([4, 2, 4, 4, 1, 3, 1])
# y is a label red or blue 7 by 1 array
y = np.array([1, 1, 1, 1, -1, -1, -1]) # 1 is red, 0 is blue

# plot the data
plt.scatter(x1, x2, c=y, s=100, cmap='bwr')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Data')
plt.show()

# print the shapes of the arrays with a description
print('x1 is a', x1.shape, 'array')
print('x2 is a', x2.shape, 'array')
```

##### (b) Sketch the optimal separating hyperplane, and provide the equa- tion for this hyperplane (of the form (9.1)).

```python
# The maximal margin hyperplane is also known as the optimal separating hyperplane
# Form of 9.1: b0 + b1*x1 + b2*x2 = 0

# Create a classifier: a support vector classifier with linear kernel
svc = sklearn.svm.SVC(kernel='linear', C=1e10)
X = np.column_stack((x1, x2))
print('The shape of X is', X.shape)
svc.fit(X,y)

# get the separating hyperplane
b1, b2 = svc.coef_[0]
b0 = svc.intercept_[0]

# to plot the hyperplane, we need to define the x1 values
x1Hyperplane = np.linspace(0, 5, 100)
x2Hyperplane = -b1/b2*x1Hyperplane - b0/b2

# plot the data
plt.plot(x1Hyperplane,x2Hyperplane) 
plt.scatter(x1, x2, c=y, s=100, cmap='bwr')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Data')

# print the hyperplane equation 
print('The hyperplane equation is ',b0,' + ',b1,'x1 + ',b2,'x2 = 0')

```

```python
#predict the output of the hyperplane
out = svc.predict(X)
print(out)


print('The output of the hyperplane is ', out)
```

##### (d) On your sketch, indicate the margin for the maximal margin hyperplane.

```python
# Compute the margin width
w = svc.coef_
margin_width = 2 / np.linalg.norm(w)
# print the margin width
print('The margin width is ', margin_width)


# Plot the data and the hyperplane and margin lines
plt.scatter(x1, x2, c=y, s=100, cmap='bwr')
plt.plot(x1Hyperplane,x2Hyperplane)
plt.plot(x1Hyperplane,x2Hyperplane + margin_width, 'k--')
plt.plot(x1Hyperplane,x2Hyperplane - margin_width, 'k--')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Data')
plt.show()
```

(e) Indicate the support vectors for the maximal margin classifier.

```python
# the support vectors are the data points that are closest to the hyperplane

# Find the support vectors
support_vectors = svc.support_vectors_
print('The support vectors are ', support_vectors)

plt.scatter (x1, x2, c=y, s=100, cmap='bwr')
# plot the first column vs the second column of the support vectors
plt.scatter(support_vectors[:,0], support_vectors[:,1], s=100, facecolors='green')
plt.plot(x1Hyperplane,x2Hyperplane)
plt.plot(x1Hyperplane,x2Hyperplane + margin_width, 'k--')
plt.plot(x1Hyperplane,x2Hyperplane - margin_width, 'k--')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Data')
# the green points are the support vectors
# label the support vectors
plt.text(2, 5, 'Support Vectors are green', fontsize=12)
plt.show()


```
